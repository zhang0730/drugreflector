```python
问题：为什么这里要生成Pseudobulk single-cell data？
```

生成 Pseudobulk 单细胞数据的核心目的是将高噪声、高维度的单细胞数据转换为低噪声、样本级的伪批量表达谱，既解决了单细胞数据的技术缺陷，又适配了 DrugReflector 模型（尤其是 Signature Refinement）对批量转录组输入的要求，最终保证药物筛选和签名优化结果的可靠性与可解释性。 

简单来说：Pseudobulk 是单细胞数据接入 DrugReflector 工具链的 “桥梁”，让单细胞数据能复用为批量数据设计的成熟分析流程。
